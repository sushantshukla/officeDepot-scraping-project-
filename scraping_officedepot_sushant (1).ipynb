```python
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import matplotlib.pyplot as plt
import pandas as pd
import re 

# importing for exceptions
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException
from selenium import webdriver
```

```python
driver=webdriver.Chrome(executable_path=r"C:\Users\Deepti Shukla\Downloads\chromedriver.exe")
```

```python
file = pd.read_csv('Sushant.csv', encoding = 'ISO-8859-1')
file['ID / SKU'].dropna(inplace = True)
ids = file['ID / SKU']
driver.get('https://www.officedepot.com/')
ids
```

```python
def text_box(search_this):
    box = driver.find_element_by_class_name('od-header-search-input') # finding text box using its class
    box.clear() # clearing it so that if some other thing is written in it then it is removed first
    box.send_keys(search_this) # searching for the keyword given
    box.submit()
```

```python
def extract_data(id) :
    manufacturer = []
    product_type = []
    brand_name = []
    manufacturer_id = []
    office_max = []
    for i in id :
        text_box(int(i))
        wait = WebDriverWait(driver, 25)
        try:
        
        
            try:
                wait.until(EC.presence_of_element_located((By.ID,'attributemanufacturerkey')))
                a = driver.find_element_by_id('attributemanufacturerkey')
                manufacturer.append(a.get_attribute('innerHTML').strip())

            except NoSuchElementException:
                manufacturer ='Not found'
                pass


            try:
                wait.until(EC.presence_of_element_located((By.ID,'attributeproduct_typekey')))
                a1 = driver.find_element_by_id('attributeproduct_typekey')
                product_type.append(a1.get_attribute('innerHTML').strip())

            except NoSuchElementException:
                product_type='Not found'
                pass


            try:

                wait.until(EC.presence_of_element_located((By.ID,'attributebrand_namekey')))
                a2 = driver.find_element_by_id('attributebrand_namekey')
                brand_name.append(a2.get_attribute('innerHTML').strip())

            except NoSuchElementException:
                brand_name ='Not found'
                pass

            try:

                wait.until(EC.presence_of_element_located((By.ID,'basicInfoManufacturerSku')))
                a3 = driver.find_element_by_id('basicInfoManufacturerSku')
                manufacturer_id.append(a3.get_attribute('innerHTML').strip().replace("&nbsp;", ""))

            except NoSuchElementException:
                manufacturer_id ='Not found'
                pass


            try:

                a4 = driver.find_element_by_xpath('//tbody[@class = "permanentRows"]//tr[2]/td[2]')
                office_max.append(a4.text)

            except NoSuchElementException:
                office_max ='Not found'
                pass
        
        except TimeoutException:
            print("Loading took too much time!")
            
    
    
    
    return manufacturer, manufacturer_id, product_type, brand_name, office_max

```

```python
extract_data(ids)
```

```python
manufacturer, manufacturer_id, product_type, brand_name, office_max=extract_data(ids)
```

```python
dataset=pd.DataFrame({'MFG NAME':manufacturer,'MFG #':manufacturer_id,'BRAND NAME': brand_name,'PRODUCT TYPE':product_type,'OFFICE MAX #':office_max})
#df = pd.DataFrame(list(zip(manufacturer, manufacturer_id, brand_name,product_type, office_max)),
                 # columns = ['MFG NAME', 'MFG #', 'BRAND NAME', 'PRODUCT TYPE', 'OFFICE MAX #'])

#df
dataset
```
